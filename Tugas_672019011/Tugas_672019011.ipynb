3. **Bar Chart**

Alasan memilih bar chart adalah karna dalam soal ini kita mencari jumlah terendah dan yang paling cocok adalah bar chart untuk mencarinya

**Penjelasan**

Yang terbawah dan harus disingkirkan dilihat dari chart adalah **Kue Wajik, Pastel dan Puding Vanilla**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

makanan=('Donat','Pastel','Kue Coklat','Ice Cream','Puding Vanila','Brownies','Puding Strawberi','Puding Coklat','Ice Cream Nutela','Kue Coklat-Keju','Kue Wajik','Kue Sus','Mochi')
penjualan=(14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'MakananPenutup' :makanan, 'DataPenjualan' : penjualan,})
df.sort_values(by='DataPenjualan', inplace = True)

kor_x = np.arange(len(df))
plt.bar(kor_x, df['DataPenjualan'], tick_label=df['MakananPenutup'])
plt.xticks(rotation=90)
plt.ylabel('DataPenjualan')
plt.title('Makanan Penutup')
plt.show()
```
