```python
import pandas as pd
import numpy as np

umur = {'umur': [29,43,np.nan,25,34,np.nan,50]}
data = pd.DataFrame(umur)
```

```python
from feature_engine.imputation import EndTailImputer

#membuat imputer
imputer = EndTailImputer(imputation_method='gaussian', tail='right')

#fitkan-imputer ke set
imputer.fit(data)

#ubah data
test_data = imputer.transform(data)

#tampilkan data
test_data
```

**Kategorikal**

```python
from sklearn.impute import SimpleImputer
mobil = {'merk':['Ford','Ford','Toyota','Honda',np.nan,'Toyota','Honda','Toyota',np.nan,np.nan]}
data = pd.DataFrame(mobil)
```

```python
data
```

```python
imp = SimpleImputer(strategy='most_frequent')
```

```python
imp.fit_transform(data)
```

**Bagaimana kalau kita butuh untuk mengatasi missing value kategorikal dan numerikal secara bersamaan?**

```python
from feature_engine.imputation import RandomSampleImputer

data = {'Jenis Kelamin': ['Pria','Wanita','Pria','Wanita','Pria','Wanita','Wanita',np.nan,'Pria',np.nan],
        'Umur':[29,np.nan,32,43,50,22,np.nan,52,np.nan,17]}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Buat Imputer
imputer = RandomSampleImputer(random_state=29)

#fit kan
imputer.fit(df)

#ubah data
testing_df = imputer.transform(df)
```
